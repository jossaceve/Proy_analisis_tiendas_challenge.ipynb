# 📊 Challenge 1 - Análisis de Tiendas

Análisis comparativo basado en:
1. Facturación total
2. Categorías más populares
3. Promedio de evaluación
4. Productos más y menos vendidos
5. Costo promedio de envío


```python
import pandas as pd
import numpy as np
```

## 📂 Carga de Datos

```python
# Si usas Google Colab con Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Ajustar la ruta si es necesario
ruta = "/content/drive/MyDrive/challenge1-data-science-latam-main/base-de-datos-challenge1-latam/"

df1 = pd.read_csv(ruta + "tienda_1.csv")
df2 = pd.read_csv(ruta + "tienda_2.csv")
df3 = pd.read_csv(ruta + "tienda_3.csv")
df4 = pd.read_csv(ruta + "tienda_4.csv")

df1["tienda"] = "Tienda 1"
df2["tienda"] = "Tienda 2"
df3["tienda"] = "Tienda 3"
df4["tienda"] = "Tienda 4"

df = pd.concat([df1, df2, df3, df4], ignore_index=True)

df.head()
```

## 1️⃣ Facturación Total por Tienda

```python
df["facturacion"] = df["precio"] * df["cantidad"]
facturacion_tienda = df.groupby("tienda")["facturacion"].sum().sort_values(ascending=False)
facturacion_tienda
```

## 2️⃣ Categorías Más Populares

```python
categorias = df.groupby(["tienda", "categoria"])["cantidad"].sum().reset_index()

categoria_top = categorias.loc[
    categorias.groupby("tienda")["cantidad"].idxmax()
]

categoria_top
```

## 3️⃣ Promedio de Evaluación

```python
promedio_evaluacion = df.groupby("tienda")["evaluacion_cliente"].mean()
promedio_evaluacion.sort_values(ascending=False)
```

## 4️⃣ Productos Más y Menos Vendidos

```python
productos = df.groupby(["tienda", "producto"])["cantidad"].sum().reset_index()

mas_vendidos = productos.loc[
    productos.groupby("tienda")["cantidad"].idxmax()
]

menos_vendidos = productos.loc[
    productos.groupby("tienda")["cantidad"].idxmin()
]

mas_vendidos, menos_vendidos
```

## 5️⃣ Costo Promedio de Envío

```python
costo_envio_promedio = df.groupby("tienda")["costo_envio"].mean()
costo_envio_promedio.sort_values()
```

## 📈 Visualización Básica

```python
import matplotlib.pyplot as plt

facturacion_tienda.plot(kind="bar", title="Facturación Total por Tienda")
plt.ylabel("Facturación")
plt.show()

promedio_evaluacion.plot(kind="bar", color="green", title="Promedio de Evaluación")
plt.ylabel("Evaluación Promedio")
plt.show()
```
